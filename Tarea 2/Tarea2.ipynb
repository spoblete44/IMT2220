```python

def gradiente_g(x, y):
    return np.array([2*x, -2*y])

def gradiente_f(x, y):
    return np.array([2*x, 2*y])
```

```python
#a)


niter = 100
lr = 0.01


init = np.array([2, -1])


approx = [init]
for iter in range(niter):
    gradient = gradiente_f(approx[-1][0], approx[-1][1])
    #print(gradient)
    approx.append(approx[-1] - lr * gradient)

approx = np.array(approx)


def f(x, y):
    return x**2 + y**2

X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))

Z = f(X, Y)

fig, ax = plt.subplots()


cp = ax.contour(X, Y, Z, levels=9)


ax.clabel(cp, cp.levels)


ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')
ax.plot(approx[:,0], approx[:,1], color='blue')

plt.show()
```

```python
#D)
niter = 100
lr = 0.01

# Initialization
init = np.array([1, 0])

# Gradient Descent
approx = [init]
for iter in range(niter):
    gradient = gradiente_g(approx[-1][0], approx[-1][1])
   # print(gradient)
    approx.append(approx[-1] - lr * gradient)

approx = np.array(approx)

# Define your function
def g(x, y):
    return x**2 - y**2

# Generate a grid of points in the x-y plane
X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))

# Compute the function values at each point in the grid
Z = g(X, Y)

# Create a new figure and axes
fig, ax = plt.subplots()

# Create a contour plot of the function
cp = ax.contour(X, Y, Z, levels=9)

# Add labels to the contour lines
ax.clabel(cp, cp.levels)

# Plot the sequence of approximations as a scatter plot
ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')
# Add line plot for the trajectory of the points
ax.plot(approx[:,0], approx[:,1], color='blue')

# Display the plot
plt.show()
```

```python
dh_dx = lambda x,y: 4*x**3 - 3*x**2 - 4*x
dh_dy = lambda x,y: np.sign(y)

# gradiente de h
def gradiente_h(x, y):
    return np.array([4*x**3 - 3*x**2 - 4*x, np.sign(y)])
```

```python




niter = 100
lr = 0.1

init = np.array([-0.01, 0.5])

approx_h = [init]
for iter in range(niter):
    gradient = np.array([dh_dx(approx_h[-1][0], approx_h[-1][1]), 
                         dh_dy(approx_h[-1][0], approx_h[-1][1])])
    approx_h.append(approx_h[-1] - lr * gradient)

approx_h = np.array(approx_h)

#print(approx_h)
def f(x, y):
    return x**4 - x**3 - 2*x**2 + np.abs(y)

X, Y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))


Z = f(X, Y)


fig, ax = plt.subplots()

cp = ax.contour(X, Y, Z, levels=9)
ax.clabel(cp, cp.levels)


ax.scatter(approx_h[:,0], approx_h[:,1], s=12, c=np.arange(approx_h.shape[0]), cmap='inferno')

ax.plot(approx_h[:,0], approx_h[:,1], color='blue')

plt.show()
```
